```python
# Combine 'Title', 'Ingredients', 'Directions' into a single text column for each recipe
similar_recipes_df['combined_features'] = similar_recipes_df['title'] + ' ' + similar_recipes_df['ner'] + ' ' + similar_recipes_df['directions']
```

```python
cleaned_features = similar_recipes_df['combined_features'].str.replace(r'["\[\]]', '')
```

```python
similar_recipes_df['cleaned_features']=cleaned_features
```

```python
similar_recipes_df['cleaned_features']
```

```python
similar_recipes_df['combined_features_lemmatized']= similar_recipes_df['cleaned_features'].apply(Definitions.lemmatize_text)
```

```python
# Use TF-IDF to vectorize the combined features
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(similar_recipes_df['combined_features_lemmatized'])

# Calculate cosine similarity between recipes
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get similar recipes based on cosine similarity
def get_similar_recipes(recipe_id, similarity_matrix, threshold=0.5):
    similar_recipes = []
    for idx, score in enumerate(similarity_matrix[recipe_id]):
        if idx != recipe_id and score >= threshold:
            similar_recipes.append((idx, score))
    return similar_recipes

# Example: Get similar recipes for Recipe 1 (adjust the recipe_id accordingly)
recipe_id = 2  # Change the recipe ID to find similar recipes for a different recipe
similar_recipes = get_similar_recipes(recipe_id, cosine_sim)

# Display similar recipes
print(f"Similar recipes for {similar_recipes_df['title'][recipe_id]}:")
for recipe in similar_recipes:
    print(f"Recipe: {similar_recipes_df['title'][recipe[0]]}, Similarity Score: {recipe[1]}")
```

```python
# Create an empty dictionary to store similarity scores for each recipe
similarities = {}

# Iterate through each recipe
for recipe_id in range(len(similar_recipes_df)):
    # Calculate similar recipes for the current recipe_id
    similar_recipes = get_similar_recipes(recipe_id, cosine_sim)
    
    # Get the title of the current recipe
    current_recipe_title = similar_recipes_df['title'][recipe_id]
    
    # Store the similar recipe titles and their similarity scores in the dictionary
    similar_recipes_with_scores = {similar_recipes_df['title'][sim_recipe[0]]: sim_recipe[1] for sim_recipe in similar_recipes}
    similarities[current_recipe_title] = similar_recipes_with_scores

# Convert the similarities dictionary to DataFrame and fill NaNs with 0
similarities_df = pd.DataFrame.from_dict(similarities).fillna(0)

# Display the DataFrame containing similarity scores
similarities_df
```

```python
# Iterate through each row of the DataFrame
for index, row in similar_recipes_df.iterrows():
    
    
    # Extract the recipe_id for the current row
    recipe_id = index
    
    # Calculate similar recipes for the current recipe_id
    similar_recipes = get_similar_recipes(recipe_id, cosine_sim)
    
    # Create a string to store similar recipes and their similarity scores
    similar_recipes_info = ""
    
    # Gather information about similar recipes for the current recipe
    for recipe in similar_recipes:
        recipe_title = similar_recipes_df['title'][recipe[0]]
        similarity_score = recipe[1]
        similar_recipes_info += f"Recipe: {recipe_title}, Similarity Score: {similarity_score}\n"
    
    # Add the information about similar recipes to the DataFrame
    similar_recipes_df.at[index, 'Similar Recipes'] = similar_recipes_info

# Display the updated DataFrame
similar_recipes_df
```

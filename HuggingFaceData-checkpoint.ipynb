```python
# finding similar recipes
# cosine similarity

# Combine 'Title', 'Ingredients', 'Directions' into a single text column for each recipe
whole_df['combined_features'] = whole_df['title'] + ' ' + whole_df['ner'] + ' ' + whole_df['directions']

# Use TF-IDF to vectorize the combined features
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(whole_df['combined_features'])

# Calculate cosine similarity between recipes
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get similar recipes based on cosine similarity
def get_similar_recipes(recipe_id, similarity_matrix, threshold=0.5):
    similar_recipes = []
    for idx, score in enumerate(similarity_matrix[recipe_id]):
        if idx != recipe_id and score >= threshold:
            similar_recipes.append((idx, score))
    return similar_recipes

# Example: Get similar recipes for Recipe 1 (adjust the recipe_id accordingly)
recipe_id = 5  # Change the recipe ID to find similar recipes for a different recipe
similar_recipes = get_similar_recipes(recipe_id, cosine_sim)

# Display similar recipes
print(f"Similar recipes for {whole_df['title'][recipe_id]}:")
for recipe in similar_recipes:
    print(f"Recipe: {whole_df['title'][recipe[0]]}, Similarity Score: {recipe[1]}")
```

```python
# Function to get similar recipe titles based on cosine similarity
def get_similar_recipe_titles(recipe_id, similarity_matrix, threshold=0.5):
    similar_recipe_indices = []
    for idx, score in enumerate(similarity_matrix[recipe_id]):
        if idx != recipe_id and score >= threshold:
            similar_recipe_indices.append(idx)
    similar_recipe_titles = [whole_df['title'][index] for index in similar_recipe_indices]
    return similar_recipe_titles

# Create a new column in the DataFrame to store similar recipe titles
whole_df['similar_recipe_titles'] = whole_df.index.map(lambda x: get_similar_recipe_titles(x, cosine_sim))

# Display the DataFrame with the new column
print(whole_df[['title', 'similar_recipe_titles']])
```

```python
def find_recipes_with_ingredients(ingredients_to_search, dataframe):
    
    # Search for recipes with all specified ingredients
    matching_recipes = whole_df[whole_df['ner'].apply(
        lambda x: all(ingredient.lower() in ' '.join(x).lower() for ingredient in ingredients_to_search))]

    matching_recipes = matching_recipes.sort_values(by=['rating', 'total_time'], ascending=[False, True])

    return matching_recipes
```

```python
def find_recipes_with_ingredients_sorted(ingredients_required, ingredients_optional, dataframe):
    
    # Define a function to calculate the count of optional ingredients in a recipe
    def count_optional_ingredients(recipe_ingredients):
        return sum(ingredient.lower() in ' '.join(recipe_ingredients).lower() for ingredient in ingredients_optional)
    
    # Search for recipes with all required ingredients
    matching_recipes = dataframe[dataframe['ner'].apply(
        lambda x: all(ingredient.lower() in ' '.join(x).lower() for ingredient in ingredients_required))].copy()
    
    # Calculate the count of optional ingredients for each recipe
    matching_recipes['optional_count'] = matching_recipes['ner'].apply(count_optional_ingredients)
    
    # Sort the recipes based on the count of optional ingredients and then by 'rating'
    matching_recipes = matching_recipes.sort_values(by=['optional_count', 'rating'], ascending=[False, False])
    
    return matching_recipes
```
